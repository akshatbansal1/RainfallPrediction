### Have to split dataset such that each location has an equal number of entries


```python
location_list= [i for i in data["Location"].unique()]
location_list
```

## Training Set

```python
data_training= pd.DataFrame()

for i in location_list:
    a=data[data['Location']==i]
    sample= a.sample(frac= 0.8)
    data_training= data_training.append(sample)

data_training.head()
```

## Testing test

```python
data_testing= data.copy() #first copy the original dataset 

X_merge = pd.merge(data, data_training, how='inner') #merges the two datasets of training and original

data_testing = data_testing.append(X_merge) #appends the original and training to the data_testing
data_testing['Duplicated']= data_testing.duplicated(keep=False) #hence marks all the duplocated rows with a True

data_testing_final = data_testing[~data_testing['Duplicated']] #selects rows which are not duplicated
del data_testing_final['Duplicated']

data_testing_final.head()   # data_testing= data_clustered - data_training
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV
```

```python
#using train test split by location
X_train= data_training[features]
y_train= data_training[target]

X_val= data_testing_final[features]
y_val= data_testing_final[target]
```

```python
# make an array of depths to choose from, say 1 to 14
depths = np.arange(1, 14)
```

```python
# define the number to try out for max leaf nodes 
leaf_nodes = [100, 150, 200, 250, 300]
```

```python
sample_leafs = np.arange(4,10)
```

```python
try_grid = [{'max_depth':depths,
              'max_leaf_nodes':leaf_nodes,
                'min_samples_leaf': sample_leafs}]
```

```python
# define your Model using GridSearchCV
DTM = GridSearchCV(DecisionTreeClassifier(), param_grid=try_grid, cv=10)
```

```python
DTM.fit(X_train,y_train)
```

```python
# find the best parameters
DTM.best_params_
```

```python
DTM.best_score_
```

```python
best_DT_model = DecisionTreeClassifier(max_leaf_nodes = 100, max_depth = 8)
```

```python
best_DT_model.fit(X_train, y_train)
```

```python
pred_val = best_DT_model.predict(X_val)
```

```python
# import evaluation metrics
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn import metrics
```

```python
# evaluate result 
print("MAE:",mean_absolute_error(y_val, pred_val))
print("Accuracy:",accuracy_score(y_val, pred_val, normalize=True, sample_weight=None))
print("Classification Report:",'\n',classification_report(y_val, pred_val))
print("Confusion Matrix:", '\n', confusion_matrix(y_val, pred_val))
```
